```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = 3.14159
```

```python
def powerperp(gamma,theta):
  beta = np.sqrt(1-1/gamma**2)
  dPdOmega = (1/(1-beta*np.cos(theta))**4)*[1-np.sin(theta)**2/(gamma**2*(1-beta*np.cos(theta))**2)] 
            # This is expression for radiated power where acceleration is perpendicular to velocity
  return dPdOmega
```

In the above, we define the function for angular power distribution where acceleration is perpendicular to velocity

```python
def powerpar(gamma,theta):
  beta = np.sqrt(1-1/gamma**2)
  #dPdOmega = np.sin(theta)**2/(gamma**2*(1-beta*np.cos(theta))**2)  This expression was given in Assignment. However its not consistent with Eq 4.101 in RL
  dPdOmega = np.sin(theta)**2/(1-beta*np.cos(theta))**6 # This expression is Eq 4.101 in RL
  return dPdOmega
```

In the above, we define the function for angular power distribution where acceleration is parallel to velocity

Below we plot the angular distribution of power for both th cases (1) Acceleration perpendicular to velocity and (2) Acceleration parallel to velocity. To do that, we try a polar plot but with Cartesian axes. For polar plot, the radial coordinate will be the expression for angular power and the angular coordinate will be the angle between the radial vector and horizontal axis, in this case x-axis. To express it in Cartesia

```python
for gamma in [5,10,100] :   # Herem we create a loop over different values of boost factor gamma
  theta = np.linspace(0, 2*pi, 100000)  # Here, we provide the range for theta for polar plot
  powpar = powerpar(gamma,theta)  # Here, we call the power function with appropriate values of boost gamma
  powpar = powpar/np.max(powpar) # Here, we normalize and scale the power by its maximum value.
  x = powpar*np.cos(theta)   # Here, we get the expressions for x coordinate of the parallel power 
  z = powpar*np.sin(theta)   # Here, we get the z coordinate for parallel power for polar plot  
  plt.figure()
  plt.plot(x,z)
  plt.xlim([0,1.2])         # Here, we set limits for proper comparitive plots
  plt.ylim([-0.12,0.12]) 
  plt.xlabel('x')
  plt.ylabel('z')
  plt.title("Relativistic beaming for boost factor $\gamma$ = "+str(gamma))
```
